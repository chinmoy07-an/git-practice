#### INSURANCE CHARGES ####

* To understand why some people pay more charges

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
print("Setup completed")
```

```python
# Let's load the data in a variable

insurance_path = "C:/Users/Dell/git-practice/resources/insurance.csv"
insurance_data = pd.read_csv(insurance_path)
```

```python
insurance_data.columns # to get the columns
```

```python
insurance_data.tail()
```

**INFO:**

* If your BMI is less than 18.5, it falls within the underweight range. 

* If your BMI is 18.5 to <25, it falls within the normal. 

* If your BMI is 25.0 to <30, it falls within the overweight range. 

* If your BMI is 30.0 or higher, it falls within the obese range.

```python
# Now let's check the correlation between BMI and charges
# Plotting a scatter plot will help since it shows how the variables depends on each other
# Ideally It is used to visualize the relationship between the two variables.

plt.figure(figsize=(16,7))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
plt.xlabel("BMI(Body mass Index)")
plt.ylabel("Charges")
plt.title("BMI vs charges relation")
```

***INSIGHTS:***

* The scatterplot above suggests that body mass index (BMI) and insurance charges are **positively correlated** 
* Customers with higher BMI typically also tend to pay more in insurance costs.

*(This pattern makes sense, since high BMI is typically associated with higher risk of chronic disease.)*

**INFO: CORRELATION WITH SCATTER PLOT**

1. If the value of y increases with the value of x, then we can say that the variables have a **positive correlation**.

2. If the value of y decreases with the value of x, then we can say that the variables have a **negative correlation**.

3. If the value of y changes randomly independent of x, then it is said to have a **zero corelation**.

```python
# To double-check the strength of this relationship, you might like to add a regression line, 
# or the line that best fits the data. We do this by changing the command to sns.regplot

# Regplot is used to plot data and a linear regression model fit

plt.figure(figsize=(16,7))
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges']) # regplot 
plt.xlabel("BMI(Body mass Index)")
plt.ylabel("Charges")
plt.title("BMI vs charges relation with regression line")

```

***COLOR-CODED SCATTER PLOTS***

* We can use scatter plots to display the relationships between (not two, but...) three variables! 

* One way of doing this is by color-coding the points.

*For instance, to understand how smoking affects( consists two values Yes and No) the relationship between BMI and insurance costs, we can color-code the points by 'smoker', and plot the other two columns ('bmi', 'charges') on the axes*

```python
plt.figure(figsize=(16,7))

#hue parameter determines which column in the data frame should be used for colour encoding
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue = insurance_data['smoker'])
plt.xlabel("BMI(Body mass Index)")
plt.ylabel("Charges")
plt.title("BMI and charges relation vs smoking effects ")
```

**INSIGHTS:**

* scatter plot shows that while nonsmokers tend to pay slightly more with increasing BMI, smokers pay MUCH more.

```python
# To further emphasize this fact, we can use the sns.lmplot command to add two regression lines, 
# corresponding to smokers and nonsmokers.

plt.figure(figsize=(16,7))
sns.lmplot(x='bmi', y='charges', hue ='smoker', data=insurance_data) # lmplot 
plt.xlabel("BMI(Body mass Index)")
plt.ylabel("Charges")
plt.title("BMI and charges relation vs smoking effects(Regression lines) ")
```

**NOTE:**

***The sns.lmplot command above works slightly differently than the commands you have learned about so far***

* Instead of setting x=insurance_data['bmi'] to select the 'bmi' column in insurance_data, we set x="bmi" to specify the name of the column only.

* Similarly, y="charges" and hue="smoker" also contain the names of columns.We specify the dataset with data=insurance_data

**INFO:**

1. regplot() performs a simple linear regression model fit and plot

2. lmplot() combines regplot() and FacetGrid.

3. The FacetGrid class helps in visualizing the distribution of one variable as well as the relationship between multiple variables separately within subsets of your dataset using multiple panels

*Usually, we use scatter plots to highlight the relationship between two continuous variables (like "bmi" and "charges").
However, we can adapt the design of the scatter plot to feature a **categorical variable (like "smoker")** on one of the main axes.We'll refer to this plot type as a **categorical scatter plot**.*

```python
# Let's use one categorical scatter plot SWARMPLOT

plt.figure(figsize=(10,6))
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges']) # swarmplot with default marker size 5
plt.xlabel("Smoking:(yes/No)")
plt.ylabel("Charges")
plt.title("charges relation vs smoking effects ")

#We can get below error due to size of markers (default value is s=5)
#let's plot another with s=4
#Or we can change the figure size
```

```python
plt.figure(figsize=(10,8)) # changes in size
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'], s=4) # swarmplot with marker size 4
plt.xlabel("Smoking:(yes/No)")
plt.ylabel("Charges")
plt.title("charges relation vs smoking effects ")

# Hence fixed
```

***INSIGHTS:***

* Plot shows on average, non-smokers are charged less than smokers

* Plot shows customers who pay the most are smokers; whereas the customers who pay the least are non-smokers.

**INFO:**

1. A strip plot is a scatter plot where one of the variables is categorical

2. Swarmplot is similar to stripplot() , but the points are adjusted (only along the categorical axis) so that they don't overlap
