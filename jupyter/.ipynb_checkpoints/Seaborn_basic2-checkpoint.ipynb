**EXAMPLE 1:**

#### US DEPARTMENT OF TRANSPORTATION ####
*Tracking Flight Delays*

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
print("Setup Complete")
```

```python
# let's load the data in a variable

flight_path = "C:/Users/Dell/git-practice/resources/flight_delays.csv"
flight_data = pd.read_csv(flight_path, index_col='Month')
```

```python
# lets review the data
flight_data 

# Since the dataset is small, we can easily print all of its contents. 
#This is done by simply writing the dataset name
```

***SOME INSIGHTS***
*Each entry shows the average arrival delay (in minutes) for a different airline and month (all in year 2015). Negative entries denote flights that (on average) tended to arrive early. For instance, the average American Airlines flight (airline code: AA) in January arrived roughly 7 minutes late, and the average Alaska Airlines flight (airline code: AS) in April arrived roughly 3 minutes early.*

```python
# Let's plot a bar chart showing average delay for NK airlines per month

plt.figure(figsize=(10,6))
plt.title("Average Flight delay for NK airlines")
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel("Arrival delay(in minutes)")
```

**INFO**: 

*  sns.barplot - This tells the notebook that we want to create a bar chart.
*  x=flight_data.index - This determines what to use on the horizontal axis
*  y=flight_data['NK'] - This sets the column in the data that will be used to determine the height of each bar

***NOTE***:

*You must select the indexing column with flight_data.index, and it is not possible to use flight_data['Month'] (which will return an error). This is because when we loaded the dataset, the "Month" column was used to index the rows. We always have to use this special notation to select the indexing column*


```python
# Let's plot a bar chart showing average delay for UA airlines per month

plt.figure(figsize=(10,6))
plt.title("Average Flight delay for NK airlines")
sns.barplot(x=flight_data.index, y=flight_data['UA'])
plt.ylabel("Arrival delay(in minutes)")
```

***Now let's visualize the flight delay patterns for all airlines***

```python
plt.figure(figsize=(14, 7))
sns.heatmap(data=flight_data, annot=True)
plt.title("Average flight delay for airlines, By month")
plt.xlabel("Airlines")
```

**INFO:**

* Bullet 1 sns.heatmap - This tells the notebook that we want to create a heatmap.
* Bullet 2 data=flight_data - This tells the notebook to use all of the entries in flight_data to create the heatmap.
* Bullet 3 annot=True - This ensures that the values for each cell appear on the chart. (Leaving this out removes the numbers from each of the cells!)

**What patterns can you detect in the table?** 

*For instance, if you look closely, the months toward the end of the year (especially months 9-11) appear relatively dark for all airlines. This suggests that airlines are better (on average) at keeping schedule during these months!*

***QUESTION 1:***

*  What is the highest average score received by PC games,for any platform?
    

```python
# We can solve it by filtering out that row and checking the highest average

ign_data.filter(like='PC', axis =0)
```

```python
# Or we can use max and idxmx functions to get max rating and its column name(index) respectively

ign_data.filter(like='PC', axis =0).max(axis=1)
```

```python
ign_data.filter(like='PC', axis =0).idxmax(axis=1)
```

```python
# or we can just create a function that returns both values(max and idxmax)

def max_and_idxmax(series):
    """Return a tuple of (max, idxmax) from a pandas.Series"""
    return series.max(axis=1), series.idxmax(axis=1)

# same for min and idxmin

def min_and_idxmin(series):
    """Return a tuple of (max, idxmax) from a pandas.Series"""
    return series.min(axis=1), series.idxmin(axis=1)
```

```python
max_and_idxmax(ign_data.filter(like='PC', axis =0)) #calling func
```

***QUESTION 2:***

* On the Playstation Vita platform, which genre has the lowest average score?

```python
# We can solve it by filterig out that row and checking the lowest 

ign_data.filter(like='PlayStation Vita', axis=0)
```

```python
# or let's just call the min_and_idxmin function on the data

min_and_idxmin(ign_data.filter(like='PlayStation Vita', axis=0))
```

**CASE STUDY:**

*Your favorite video game has been Mario Kart Wii, a racing game released for the Wii platform*

*you're considering creating your very own racing game for the Wii platform*

*so you want to check average score for racing games, for each platform*

```python
# let plot a bar graph to see the average score across all platforms for racing

plt.figure(figsize=(16,7))
sns.barplot(x=ign_data.index, y=ign_data['Racing'])
plt.title("Average rating of Racing games across Platforms")
plt.xlabel("Platforms")
plt.ylabel("Ratings for Racing")

# Now you see we are unable to check the platforms names properly
# One way to fix it is using axes  and set labels
```

```python
plt.figure(figsize=(16,7))
ax = sns.barplot(x=ign_data.index, y=ign_data['Racing']) # using ax as in axes
plt.title("Average rating of Racing games across Platforms")
ax.set_xticklabels(ign_data.index, rotation=90) # setting label as rotation 90
plt.xlabel("Platforms")
plt.ylabel("Ratings for Racing")
```
